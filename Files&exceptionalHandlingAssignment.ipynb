#Theory Questions

1. What is the difference between interpreted and compiled languages?
  - Compiled languages translate code to machine code before execution, making them faster. Interpreted languages execute code line-by-line using an interpreter, making them slower but easier to debug and test.

2. What is exception handling in Python?
  - Exception handling in Python is a mechanism to handle runtime errors. It allows us to catch and handle errors, and help program from crashing.

3. What is the purpose of the finally block in exception handling?
  - Purpose: The finally block is used to specify code that should be executed regardless of whether an exception was raised or not.

4. What is logging in Python?
  - Logging in Python is a mechanism to record events that occur during the execution of a program.

5. What is the significance of the __del__ method in Python?
  - The __del__ method is called when an object is about to be destroyed.

6. What is the difference between import and from ... import in Python?
  - Import loads the entire module, while from ... import only loads the specified attributes.

7. How can you handle multiple exceptions in Python?
  - We can handle multiple exceptions using a try-except block.

8. What is the purpose of the with statement when handling files in Python?
  - Purpose of the with statement is to automatically close the file after use.

9. What is the difference between multithreading and multiprocessing?
  - Multithreading runs multiple threads within a single process, while multiprocessing runs multiple processes.

10. What are the advantages of using logging in a program?
  - Advantages include tracking errors, debugging, and performance.

11. What is memory management in Python?
  - Memory management in Python is the process of managing the allocation and deallocation of memory. Python uses a garbage collector to manage memory.

12. What are the basic steps involved in exception handling in Python?
  - Basic steps include catching and handling exceptions and is done using try-except blocks.

13. Why is memory management important in Python?
  - Memory management is important because it helps manage the allocation and deallocation of memory.

14. What is the role of try and except in exception handling?
  - Try and except in exception handling is used to catch and handle exceptions.

15. How does Python's garbage collection system work?
  - Python's garbage collection system is used to manage memory. It frees up memory that is no longer in use.

16. What is the purpose of the else block in exception handling?
  - Purpose of the else block is to specify code that should be executed if no exceptions are raised.

17. What are the common logging levels in Python?
  - Common logging levels in Python are INFO, WARNING, ERROR, and CRITICAL.

18. What is the difference between os.fork() and multiprocessing in Python?
  - os.fork() creates a new process, while multiprocessing creates multiple processes.

19. What is the importance of closing a file in Python?
  - Closing a file in Python is important to free up system resources.

20. What is the difference between file.read() and file.readline() in Python?
  - file.read() reads the entire file, while file.readline() reads one line at a time.

21. What is the logging module in Python used for?
  - The logging module in Python is used to log events during program execution.

22. What is the os module in Python used for in file handling?
  - The os module in Python is used to interact with the operating system.

23. What are the challenges associated with memory management in Python?
  - Challenges associated with memory management in Python include memory leaks and memory fragmentation.

24.  How do you raise an exception manually in Python?
   - Using the raise keyword we can raise an exception manually in Python.

25. Why is it important to use multithreading in certain applications?
